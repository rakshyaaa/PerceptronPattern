```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
def plot_data(X, Y, title):
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Set1)
    plt.title(title)
    plt.legend()
    plt.show()
```

1. Data Generation

```python
#Dataset with noise=0.20
X1, y1 = make_moons(n_samples=250, noise=0.20, random_state=51)
plot_data(X1,y1,"Noise=0.20")
```

```python
#Dataset with noise=0.35
X2, y2 = make_moons(n_samples=250, noise=0.35, random_state=51)
plot_data(X2,y2,"Noise=0.35")
```

2. kNN Classifier (Noise = 0.20)

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

#Splitting
X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.25, random_state=509)


k_values = range(3, 11)
accuracies = []

# Training and evaluating the kNN model for different k values
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

# Plotting the results
plt.figure()
plt.plot(k_values, accuracies, marker='o')
plt.title("kNN Accuracy vs Number of Neighbors (k)")
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Accuracy")
plt.grid(True)
plt.show()
```
