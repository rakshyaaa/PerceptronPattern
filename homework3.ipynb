```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
def plot_data(X, Y, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Set1, edgecolor='k')
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
    plt.show()
```

1. Data Generation

```python
#Dataset with noise=0.20
X1, y1 = make_moons(n_samples=250, noise=0.20, random_state=51)
plot_data(X1,y1,"Noise=0.20")
```

```python
#Dataset with noise=0.35
X2, y2 = make_moons(n_samples=250, noise=0.35, random_state=51)
plot_data(X2,y2,"Noise=0.35")
```

2. kNN Classifier (Noise = 0.20)

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

#Splitting
X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.25, random_state=509)

#Trainning kNN
k = 3
knn = KNeighborsClassifier(n_neighbors = k)
knn.fit(X_train, y_train)

#Making predicitons on the test set
y_pred = knn.predict(X_test)

# Evaluating the model 
accuracy = accuracy_score(y_test, y_pred) 
print(f"Knn 3 Accuracy: {accuracy:.2f}") 
```

3. Experiment with noisy data set 

KNN with noisy data of 0.35

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

#Splitting
X_train, X_test, y_train, y_test = train_test_split(X2, y2, test_size=0.25, random_state=509)

#Trainning kNN
k = 3
knn = KNeighborsClassifier(n_neighbors = k)
knn.fit(X_train, y_train)

#Making predicitons on the test set
y_pred = knn.predict(X_test)

# Evaluating the model 
accuracy = accuracy_score(y_test, y_pred) 
print(f"Knn 3 Accuracy: {accuracy:.2f}") 
```
