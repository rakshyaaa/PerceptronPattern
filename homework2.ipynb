```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
def plot_data(X, Y, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Set1, edgecolor='k')
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
    plt.show()
```

1. Data Generation

```python
#Dataset with noise=0.20
X1, y1 = make_moons(n_samples=250, noise=0.20, random_state=51)
plot_data(X1,y1,"Noise=0.20")
```

```python
#Dataset with noise=0.35
X2, y2 = make_moons(n_samples=250, noise=0.35, random_state=51)
plot_data(X2,y2,"Noise=0.35")
```

2. Building a decision tree

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

#Train-test split
X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.25, random_state=509) #25% of data used for testing and 75% for training

# Build and train decision tree
dt_classifier = DecisionTreeClassifier(random_state=509)
dt_classifier.fit(X_train, y_train)

# Predict and evaluate
y_pred = dt_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Base DT Accuracy: {accuracy:.2f}")
```

3. Experimenting with decision tree depth

```python
for depth in range(1,10):
    dt = DecisionTreeClassifier(max_depth=depth, random_state=509)
    X_train, X_test, y_train, y_test = train_test_split(X2, y2, test_size=0.25, random_state=509)
    dt.fit(X_train, y_train)
    y_pred = dt.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"Depth={depth}, Accuracy={acc:.2f}")
```
